# prÃ©traitement les donnÃ©es 

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Encodage de la variable 'Gender'
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])

# Normalisation des variables numÃ©riques
scaler = StandardScaler()
numeric_columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Afficher les premiÃ¨res lignes du DataFrame aprÃ¨s le prÃ©traitement
print(df.head())
```

# Choix du nombre de clusters (K) 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# CrÃ©er une liste vide pour stocker les valeurs d'inertie et de coefficient de silhouette
inertia_values = []
silhouette_scores = []

# DÃ©finir une plage de valeurs de K
k_range = range(2, 11)

# Boucle sur les diffÃ©rentes valeurs de K
for k in k_range:
    # CrÃ©er le modÃ¨le K-means
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    
    # Calculer l'inertie et l'ajouter Ã  la liste
    inertia_values.append(kmeans.inertia_)
    
    # Calculer le coefficient de silhouette et l'ajouter Ã  la liste
    silhouette_scores.append(silhouette_score(df, kmeans.labels_))

# Tracer le graphique de l'inertie en fonction de K
plt.figure(figsize=(10, 5))
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Nombre de clusters (K)')
plt.ylabel('Inertie')
plt.title('MÃ©thode du coude')
plt.show()

# Tracer le graphique du coefficient de silhouette en fonction de K
plt.figure(figsize=(8, 5))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Nombre de clusters (K)')
plt.ylabel('Coefficient de silhouette')
plt.title('Coefficient de silhouette')
plt.show()
```

```python
# from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# DÃ©finir le nombre de clusters (K)
k = 4

# SÃ©lectionner les caractÃ©ristiques Ã  utiliser
features = ['Age', 'Spending Score (1-100)']
X = df[features]

# CrÃ©er le modÃ¨le K-means avec le nombre de clusters spÃ©cifiÃ©
kmeans = KMeans(n_clusters=k, random_state=42)

# EntraÃ®ner le modÃ¨le sur les donnÃ©es
kmeans.fit(X)

# DÃ©terminer les limites de la grille
x_min, x_max = X['Age'].min() - 1, X['Age'].max() + 1
y_min, y_max = X['Spending Score (1-100)'].min() - 1, X['Spending Score (1-100)'].max() + 1

# GÃ©nÃ©rer une grille de points pour la visualisation
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# PrÃ©dire les clusters pour chaque point de la grille
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Mettre en forme les donnÃ©es pour la visualisation
Z = Z.reshape(xx.shape)

# Afficher le graphique avec les rÃ©gions colorÃ©es par cluster
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X['Age'], X['Spending Score (1-100)'], c=kmeans.labels_, s=20, cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters K-means')
plt.colorbar()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# SÃ©lection des colonnes pour le clustering
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values

# DÃ©finition du nombre de clusters
k = 4

# CrÃ©ation du modÃ¨le K-means
kmeans = KMeans(n_clusters=k, random_state=42)

# EntraÃ®nement du modÃ¨le sur les donnÃ©es
kmeans.fit(X)

# Ajout des informations sur les clusters au DataFrame
df['Cluster'] = kmeans.labels_

# Affichage des centres des clusters
print("Centroids:")
print(kmeans.cluster_centers_)

# Visualisation des clusters
plt.figure(figsize=(10, 6))

# Parcourez chaque cluster et affichez les points correspondants
for cluster in range(k):
    plt.scatter(X[df['Cluster'] == cluster, 0], X[df['Cluster'] == cluster, 1], label=f'Cluster {cluster}')

# Affichage des centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')

plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title('Clustering des clients')
plt.legend()
plt.show()
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.cluster import KMeans

# Importation des donnÃ©es
df = pd.read_csv('Mall_Customers.csv')

# SÃ©lection des caractÃ©ristiques pour le clustering
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values

# Application de la mÃ©thode du coude pour dÃ©terminer le nombre optimal de clusters
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, init='k-means++', random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# TracÃ© du graphique de la mÃ©thode du coude
plt.plot(range(1, 11), inertia, marker='o')
plt.title('MÃ©thode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
plt.show()

# Application du modÃ¨le K-means avec le nombre optimal de clusters
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Visualisation des clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters des clients')
plt.xlabel('Age')
plt.ylabel('Revenu annuel (k$)')
plt.legend()
plt.show()
```
